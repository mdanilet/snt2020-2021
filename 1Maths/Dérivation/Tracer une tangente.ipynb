# Tracer des tangentes à la courbe grâce à la dérivation
## Implémenter la fonction
On considère la fonction définie sur $\mathbf{R}$ par $f(x)=3x^3+2x^2-7x+12$

Complétez le script pour implémenter la fonction :

```python
def f(x):
    return 3*x**3+2*x**2-7*x+12
```

## Tracer sa courbe
Python ne suffit pas pour tracer une courbe, il faut également importer le module `matplotlib` :

```python
from matplotlib import pyplot as plt

def courbe(fonction, a, b):
    plt.plot([f(x) for x in range(a,b)])
    
courbe(f, 0,10)
plt.show()

```

## Valeur approchée d'un nombre dérivé
On rappelle la formule du taux d'accroissement entre $a$ et $a+h$ :
$\tau = \frac {f(a+h)-f(a)} {h}$

Le moteur de Python ne sait pas calculer une limite. En revanche, on peut lui préciser une valeur de $h$ très petite :

```python
def nbderive(fonction, a):
    h=10**(-10) #10 puissance -10
    return (fonction(a+h)-fonction(a))/h
```

## Equation de la tangente
On souhaite ajouter au graphique précédent la tangente à la courbe au point d'abscisse 6.
Complétez la fonction yT qui à x associe `y=mx+p`

```python
def yT(fonction, a, x):
    yA=fonction(a)
    m=nbderive(fonction, a)
    p=yA-m*a
    return m*x+p
def traceTangente(fonction, a, fin):
    plt.plot([yT(fonction, a, x) for x in range(0, fin)])

fin=5
#courbe(f, 0, fin)
for i in range(fin):
    traceTangente(f, i, fin)
```
